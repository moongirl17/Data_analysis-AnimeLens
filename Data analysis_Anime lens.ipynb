```python
df=df[~np.isnan(df["rating"])]

df['genre'] = df['genre'].fillna(
df['genre'].dropna().mode().values[0])

df['type'] = df['type'].fillna(
df['type'].dropna().mode().values[0])

df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
dftt = df.drop(columns=['Unnamed: 0', 'type', 'episodes', 'members'])
```

```python
dftt.head()
```

```python
dftt["genre"] = dftt["genre"].str.replace(", ", ",")

# Then convert the genre column into a list so you can hot one encode the genres.
dftt["genre"] = dftt["genre"].str.split(",")

dftt.head()
```

```python
# Using scikit learn's MLB package to one hot encode the genres
from sklearn.preprocessing import MultiLabelBinarizer

# Code from https://stackoverflow.com/questions/45312377/how-to-one-hot-encode-from-a-pandas-column-containing-a-list
mlb = MultiLabelBinarizer(sparse_output=True)

dftt = dftt.join(pd.DataFrame.sparse.from_spmatrix(
                mlb.fit_transform(dftt["genre"]),
                index=dftt.index,
                columns=mlb.classes_))

# Drop the origininal genre column
dftt.drop("genre", axis=1, inplace=True)
```
